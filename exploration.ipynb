```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
inputfile = 'data/merged/allschools.csv'
```

```python
df = pd.read_csv(inputfile, dtype={'Estimate!!Households!!Median income (dollars)': int})
```

```python
df.loc[df['Adm'].isna(), 'Adm'] = 0
df['Ratio'] = df['Adm'] / df['App']
```

```python
alldf = df.loc[(df['Ratio'] != 0) & (df['Ratio'] != 1)]
privatedf = df.loc[(df['Ratio'] != 0) & (df['Ratio'] != 1) & (df['Public Yes/No'] == 0)]
publicdf = df.loc[(df['Ratio'] != 0) & (df['Ratio'] != 1) & (df['Public Yes/No'] == 1)]
```

```python
X = np.c_[alldf['Median income (dollars)']]
y = np.c_[alldf['Ratio']]

# Plot scatterplot 
plt.scatter(X, y, alpha=.5, s=alldf['App']/10)

plt.xlabel('Median Income')
plt.ylabel('Ratio Admitted')
plt.title('Acceptance vs Median Income')

plt.show()
```

```python
X = np.c_[publicdf['Median income (dollars)']]
y = np.c_[publicdf['Ratio']]

# Plot scatterplot 
plt.scatter(X, y, alpha=.5, s=publicdf['App']/10)
plt.legend(loc='best')

plt.xlabel('Median Income')
plt.ylabel('Ratio Admitted')
plt.title('Public Acceptance vs Median Income')

plt.show()
```

```python
X = np.c_[privatedf['Median income (dollars)']]
y = np.c_[privatedf['Ratio']]

# Plot scatterplot 
plt.scatter(X, y, alpha=.5, s=privatedf['App']/5)
plt.legend(loc='best')

plt.xlabel('Median Income')
plt.ylabel('Ratio Admitted')
plt.title('Private Acceptance vs Median Income')

plt.show()
```
